```python
# which country has highest population
df[df['population'] == df['population'].max()]['country']
df.columns
```

```python
# what is capital of country with highest population
df[df['population'] == df['population'].max()]['capital_city']
```

```python
# which country has the least population
df[df['population'] == df['population'].min()]['country']
```

```python
# top 5 countries with highest democratic score

df.sort_values(by = 'democracy_score', ascending=False, inplace=True)
```

```python
df['country'].head()
```

```python
# how many total regions are there

df['region'].value_counts().count()
```

```python
# how many countries lies in the eastern europe region

df['region'].value_counts()['Eastern Europe']
```

```python
# how many countries are there in eastern europe region with name of country

df[df['region'] == 'Eastern Europe']['country']
```

```python
# who is political leader of 2nd highest populated country

df[df['population'] == df['population'].nlargest(5).iloc[3]]['political_leader']
```

```python
# how many countries are there whose political leader are unknown
df[df['political_leader'].isna()]['country'].count()
```

```python
# how many country have republic in their name
count = 0
def counting(txt):
    global count
    if 'republic' in txt.lower():
        count += 1
    return txt

df['country_long'] = df['country_long'].apply(counting)
print(count)
```

```python
# how many countries in african region has highest population

africa_df = df[df['continent'] == 'Africa']
africa_df[africa_df['population'] == africa_df['population'].max()]['country']
```
